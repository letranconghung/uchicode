```python
vals = list(range(1, 14))
vals
```

```python
sum_values = []
for i in vals:
    for j in vals:
        for k in vals:
            for l in vals:
                sum_values.append(i + j + k + l)

import matplotlib.pyplot as plt

import pandas as pd

plt.figure(figsize=(16, 9))
sum_values_series = pd.Series(sum_values)
ax = sum_values_series.value_counts().sort_index().plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.xlabel('Sum Values')
plt.ylabel('Frequency')
plt.title('Frequency of Sum Values')
plt.show()

std_dev = sum_values_series.std()
std_dev
```
